```python
import tkinter as tk
from tkinter import messagebox

def calculate_eor():
    try:
        # Ambil input dari user
        sor = float(entry_sor.get())  # Saturasi minyak sisa (residual oil saturation)
        vis_oil = float(entry_vis_oil.get())  # Viskositas minyak (cP)
        vis_inj = float(entry_vis_inj.get())  # Viskositas fluida injeksi (cP)
        efficiency = float(entry_efficiency.get())  # Efisiensi injeksi (%)

        # Perhitungan faktor perolehan (Recovery Factor)
        mobility_ratio = vis_inj / vis_oil
        rf = efficiency * (1 - sor) / (1 + mobility_ratio)

        # Tampilkan hasil
        result_label.config(text=f"Recovery Factor: {rf:.2f}%")
    except ValueError:
        messagebox.showerror("Input Error", "Masukkan nilai numerik yang valid!")

# Buat jendela utama
root = tk.Tk()
root.title("Kalkulator Enhanced Oil Recovery (EOR)")
root.geometry("400x300")

# Judul
title_label = tk.Label(root, text="Kalkulator EOR", font=("Arial", 16))
title_label.pack(pady=10)

# Input Saturasi Minyak Sisa (Sor)
frame_sor = tk.Frame(root)
frame_sor.pack(pady=5)
label_sor = tk.Label(frame_sor, text="Saturasi Minyak Sisa (Sor):")
label_sor.pack(side=tk.LEFT)
entry_sor = tk.Entry(frame_sor)
entry_sor.pack(side=tk.LEFT, padx=5)

# Input Viskositas Minyak
frame_vis_oil = tk.Frame(root)
frame_vis_oil.pack(pady=5)
label_vis_oil = tk.Label(frame_vis_oil, text="Viskositas Minyak (cP):")
label_vis_oil.pack(side=tk.LEFT)
entry_vis_oil = tk.Entry(frame_vis_oil)
entry_vis_oil.pack(side=tk.LEFT, padx=5)

# Input Viskositas Fluida Injeksi
frame_vis_inj = tk.Frame(root)
frame_vis_inj.pack(pady=5)
label_vis_inj = tk.Label(frame_vis_inj, text="Viskositas Fluida Injeksi (cP):")
label_vis_inj.pack(side=tk.LEFT)
entry_vis_inj = tk.Entry(frame_vis_inj)
entry_vis_inj.pack(side=tk.LEFT, padx=5)

# Input Efisiensi Injeksi
frame_efficiency = tk.Frame(root)
frame_efficiency.pack(pady=5)
label_efficiency = tk.Label(frame_efficiency, text="Efisiensi Injeksi (%):")
label_efficiency.pack(side=tk.LEFT)
entry_efficiency = tk.Entry(frame_efficiency)
entry_efficiency.pack(side=tk.LEFT, padx=5)

# Tombol Hitung
calculate_button = tk.Button(root, text="Hitung", command=calculate_eor)
calculate_button.pack(pady=10)

# Label Hasil
result_label = tk.Label(root, text="", font=("Arial", 14))
result_label.pack(pady=10)

# Jalankan aplikasi
root.mainloop()
```
